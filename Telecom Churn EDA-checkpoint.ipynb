### Analysing Statewise Churn

```python
plt.figure(figsize=(28,6))
sns.countplot(data,x="State",hue="Churn")
plt.title("Churn Vs State",fontsize=15)
```

```python
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = (18, 7)
s1=df['State'].unique()
s2=df.groupby(['State'])['Churn'].mean()
plt.plot(s1,s2,color="c",marker="o",linewidth=2,markersize=12)
plt.axhline(0.10,color="r",ls="-.",linewidth=3)
plt.title(" States churn rate", fontsize = 20,color="c")
plt.xlabel('state', fontsize = 15)
plt.ylabel('churn rate', fontsize = 15)

plt.show()
```

### Insight 4:
By Comparing with State and Churn percentage there is more churns from specific states

```python
a=pd.DataFrame(((df.groupby(["State"])["Churn"].mean())*100).sort_values(ascending=False))

a.plot.bar(color="c")
plt.axhline(10,color="r",ls="-.",linewidth=3)
plt.title("State wise Churn percentage",fontsize=25,color="c")
plt.xlabel("States",fontsize=15)
plt.ylabel("Churn Percentage",fontsize=15)
plt.show()
```

```python
((df.groupby(['State'])['Churn'].mean())*100).sort_values(ascending = False).head(10).plot.bar(color = ['violet','indigo','b','g','y','orange','r'])
```
